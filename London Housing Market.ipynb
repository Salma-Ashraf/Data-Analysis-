## Cleaning Data

``(1) Converting date column to "date time" data type ``

```python
data['date'] = pd.to_datetime(data.date)
```

```python
data.dtypes
```

``(2) Creating a new column containing the year ``

```python
data["year"] = data['date'].dt.year
```
